```python
#selected_data = pd.DataFrame(final['name'],final['selected_option'], final['topic_id'], final['time'])

# Select the desired columns from the existing DataFrame 'final'
selected_data = final[['name', 'selected_option', 'topic_id', 'time']]

# Display the new DataFrame
print(selected_data)
```

```python
selected_data
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Extract the 'count' column for clustering
X = selected_data[['time', 'topic_id']]

# Define the number of clusters
num_clusters = 3

# Initialize the KMeans model
kmeans = KMeans(n_clusters=num_clusters, random_state=0)

# Fit the model
kmeans.fit(X)

# Assign the cluster labels to the original DataFrame
final['cluster'] = kmeans.labels_

# Display the DataFrame with cluster labels
print(final)
```

```python
final.head(50)
```

```python
# Plot the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=final, x='name', y='time', hue='cluster', palette='viridis', s=100)
plt.title('K-means Clustering of time/topic_id')
plt.xlabel('User ID')
plt.ylabel('time taken')
plt.legend(title='Cluster')
plt.show()
```
