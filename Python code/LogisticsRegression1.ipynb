```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import math
%matplotlib inline
df=pd.read_csv('titanic.csv')
df.head(10)
```

```python
df.drop(['name','home.dest','boat','body'], axis=1,inplace=True)
df.head(10)
```

```python
df.drop('cabin',axis=1,inplace=True)
df.head(5)
```

```python
df.drop('ticket',axis=1,inplace=True)
df.head(5)
```

```python
df['sex'].unique()
```

```python
#nan is removed from row
df.dropna(inplace=True)
df.count()
```

```python
df.isnull().sum()
```

```python
df['sex'].unique()
```

```python
vsex=pd.get_dummies(df['sex'],drop_first=True)
vsex.head(5)
```

```python
df=pd.concat((df,vsex),axis=1)
df.head(5)
```

```python
df.drop(['sex'],axis=1,inplace=True)
df.head(5)
```

```python
df['embarked'].unique()
```

```python
#coverting variable to quantity data
vembarked=pd.get_dummies(df['embarked'],drop_first=True)
vembarked.head(5)
```

```python
df=pd.concat((df,vembarked),axis=1)
df.head(5)
```

```python
df.drop(['embarked'],axis=1,inplace=True)
df.head(5)
```

```python
df.info()
```

```python
#train data
from sklearn.model_selection import train_test_split
x=df.drop('survived',axis=1)
y=df['survived']
#from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression(solver='lbfgs',max_iter=1000)
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
#print(predictions,':',y_test)
#predictions
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
res=confusion_matrix(y_test,predictions)
res
tn,tp,fn,fp=confusion_matrix(y_test,predictions).ravel()
tn,tp,fn,fp
```
