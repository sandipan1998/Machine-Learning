```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# The below 3 lines will create 3 list of values
company=["IBM","TCS","CTS","INFOSYS"]
revenue=[10,89,40,56]
profit=[5,20,12,10]

xpos=np.arange(len(company))
#xticks is list of positions where ticks need to be given.... [0,1,2,3,] in this case
plt.xticks(xpos,company)
plt.ylabel("Revenue")
plt.title("Company revenue")
#plt.bar(xpos,revenue,label="Revenue")
#plt.bar(xpos,profit,label="profit")
plt.bar(xpos-0.2,revenue,width=0.2,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.2,label="profit")
plt.legend()
plt.show()
```

```python
plt.yticks(xpos,company)
plt.ylabel("company")
plt.title("Company revenue")
#plt.barh(xpos,revenue,label="Revenue")
#plt.barh(xpos,profit,label="profit")
plt.barh(xpos-0.2,revenue,height=0.2,label="Revenue")
plt.barh(xpos+0.2,profit,height=0.2,label="profit")
plt.legend()
plt.show()
```
