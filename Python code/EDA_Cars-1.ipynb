```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cardf=pd.read_csv("Automobile_data.csv")
cardf.head(5)
```

```python
#settings to disply all columns
pd.set_option("display.max_columns",None)
cardf.head(5)
```

```python
cardf.info()
```

```python
cardf.describe()
```

```python
cardf.isnull().sum()
```

```python
#Data that contains '?' is to be replace with NaN
cardf.replace('?',np.NaN,inplace=True)
```

```python
cardf.isnull().sum()
```

```python
#fill missing data of normalized-losses,price,horsepower,peak-rpm,bore,stroke with respect
#fill misssing data catagory number of doors with the mode of the column i.e. Four

df_temp=cardf[cardf['normalized-losses']!=np.NaN]
avg1=df_temp['normalized-losses'].astype(float).mean()
cardf['normalized-losses']=cardf['normalized-losses'].replace(np.NaN,avg1).astype(float)
#cardf['normalized-losses'].isnull().sum()

df_temp=cardf[cardf['price']!=np.NaN]
avg1=df_temp['price'].astype(float).mean()
cardf['price']=cardf['price'].replace(np.NaN,avg1).astype(float)
cardf['price'].isnull().sum()

df_temp=cardf[cardf['horsepower']!=np.NaN]
avg1=df_temp['horsepower'].astype(float).mean()
cardf['horsepower']=cardf['horsepower'].replace(np.NaN,avg1).astype(float)
cardf['horsepower'].isnull().sum()

df_temp=cardf[cardf['bore']!=np.NaN]
avg1=df_temp['bore'].astype(float).mean()
cardf['bore']=cardf['bore'].replace(np.NaN,avg1).astype(float)
cardf['bore'].isnull().sum()

df_temp=cardf[cardf['peak-rpm']!=np.NaN]
avg1=df_temp['peak-rpm'].astype(float).mean()
cardf['peak-rpm']=cardf['peak-rpm'].replace(np.NaN,avg1).astype(float)
cardf['peak-rpm'].isnull().sum()

df_temp=cardf[cardf['stroke']!=np.NaN]
avg1=df_temp['stroke'].astype(float).mean()
cardf['stroke']=cardf['stroke'].replace(np.NaN,avg1).astype(float)
cardf['stroke'].isnull().sum()

cardf['num-of-doors']=cardf['num-of-doors'].replace(np.NaN,'four')
cardf.head()
```

```python
#Univeriable analysis
cardf[['engine-size','price','horsepower','bore','peak-rpm']].hist(figsize=(10,8),bins=6,color='g')
#plt.figure(figsize=(10,8))
#plt.tight_Layout()
plt.show()
```

```python
#findings

#most of the car has a curb weight is range 1900 to 3100
#The Engine size is inrange 60 to 190
#Most vechile has horsepower 50-125
#Most of the vechile are inprice range 5000 to 18000
#peak rpm is mostly distributed between 4600 to 5700
```

```python
plt.figure(1)
plt.subplot(221)
cardf['engine-type'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='red')
plt.title("Number of Engine Type frequency diagram")
plt.ylabel('Number of Engine Type')
plt.xlabel('engine-type');


plt.subplot(222)
cardf['num-of-doors'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='green')
plt.title("Number of Door frequency diagram")
plt.ylabel('Number of Doors')
plt.xlabel('num-of-doors');

plt.subplot(223)
cardf['fuel-type'].value_counts(normalize= True).plot(figsize=(10,8),kind='bar',color='purple')
plt.title("Number of Fuel Type frequency diagram")
plt.ylabel('Number of vehicles')
plt.xlabel('fuel-type');

plt.subplot(224)
cardf['body-style'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='orange')
plt.title("Number of Body Style frequency diagram")
plt.ylabel('Number of vehicles')
plt.xlabel('body-style');
plt.tight_layout()
plt.show()
```

```python
# More than 70% of vechile has ohc type of engine
#57% of the cars has 4 doors
#Gas is preferred by 85% of the vechile
#Most produced vechile are of body style sedan around 48% followed by hatchback32%
```
